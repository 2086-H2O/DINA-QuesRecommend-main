```python
import pandas as pd

def generate_q_matrix(input_txt, output_excel):
    """ç›´æ¥ç”ŸæˆQçŸ©é˜µçš„ç®€æ´å®ç°"""
    # è¯»å–åŸå§‹æ•°æ®
    with open(input_txt, 'r', encoding='utf-8') as f:
        questions = [line.strip().split() for line in f if line.strip()]

    # è‡ªåŠ¨æå–å”¯ä¸€çŸ¥è¯†ç‚¹ï¼ˆåˆ—ï¼‰
    all_concepts = sorted(list({concept for q in questions for concept in q}))
    
    # æ„å»ºQçŸ©é˜µ
    q_matrix = []
    for q in questions:
        q_matrix.append([1 if concept in q else 0 for concept in all_concepts])
    
    # è½¬æ¢ä¸ºDataFrameå¹¶è¾“å‡º
    pd.DataFrame(
        q_matrix,
        columns=all_concepts,
        index=[f"Q{i+1}" for i in range(len(q_matrix))]
    ).to_excel(output_excel)

# ä½¿ç”¨ç¤ºä¾‹
generate_q_matrix("Deepseek.txt", "output_matrix.xlsx")
```
